```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = "https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.shape
```

```python
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].unique()
```

```python
# This code will gives all cars and their type who not have clean fuel eligility
df_not_clean_fuel = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] != 'Clean Alternative Fuel Vehicle Eligible']

print('All the cars and their types that do not qualify for clean alternative fuel vehicle')
df_not_clean_fuel[['Model','Electric Vehicle Type']]
```

#### Insights 1.) Total 69892 cars that do not qualify for clean alternative fuel vehicle

```python
print('All TESLA cars with the model year, and model type made in Bothell City.')

df_tesla_bothell = df[(df['Make'] == 'TESLA') & (df['City'] == 'Bothell')][['Make','Model Year', 'Model']]
df_tesla_bothell
```

#### Insights 2.) 2407 numbers of tesla car are made in bothell city.

```python
print('All the cars that have an electric range of more than 100, and were made after 2015')
cars_electric_range_more_than_100 = df[(df['Electric Range'] > 100) & (df['Model Year'] > 2015)]
cars_electric_range_more_than_100
```

#### Insights 3.) There are 34066 cars have electric range more than 100 and built after 2015

```python
print("Plots to show the distribution between city and electric vehicle type")

city_ev_counts = df.groupby(['City', 'Electric Vehicle Type']).size().unstack().fillna(0)
city_ev_counts.plot(kind='line', stacked=True)
plt.xlabel('City')
plt.xticks(rotation='90')
plt.ylabel('Count')
plt.title('Distribution of Electric Vehicle Types by City')
plt.show()
```

#### Insights 4.) Plots to show the distribution between city and electric vehicle type
