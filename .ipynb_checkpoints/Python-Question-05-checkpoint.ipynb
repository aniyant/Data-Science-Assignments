```python
Q.5) Write a program to download the data from the given API link and then extract the following data with
    proper formatting

Link - http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes

Note - Write proper code comments wherever needed for the code understanding

Sample Data -

Excepted Output Data Attributes -
● id - int url - string
● name - string season
● - int number - int
● type - string airdate -
● date format airtime -
● 12-hour time format
● runtime - float
● average rating - float
● summary - string
● without html tags
● medium image link - string
● Original image link - string
```

```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd

# Function to download data from the provided API link
def download_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception("Failed to download data")

# Function to extract and format the required data attributes
def extract_data(data):
    show_data = {
        'id': data['id'],
        'url': data['url'],
        'name': data['name']
    }

    episodes = data['_embedded']['episodes']
    formatted_episodes = []
 
    for episode in episodes:
        formatted_episode = {
            'id' : episode['id'],
            'name' : episode['name'],
            'url' : episode['url'],
            'season': episode['season'],
            'number': episode['number'],
            'type': episode['type'],
            'airdate': episode['airdate'],
            'airtime': episode['airtime'],
            'runtime': episode['runtime'],
            'average_rating': episode['rating']['average'],
            'summary': clean_html_tags(episode['summary']),
            'medium_image': episode['image']['medium'],
            'original_image': episode['image']['original']
        }
        formatted_episodes.append(formatted_episode)

    return show_data, formatted_episodes

# Function to clean HTML tags from the summary text
def clean_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    cleaned_text = soup.get_text()
    return cleaned_text

# Url to download data from
url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"


# Calling download function to download data
data = download_data(url)



# Extract and format the required data attributes
show_info, episodes_info = extract_data(data)

# Print the show information
print("Show Information:")
print("Name of the show",show_info['name'])
print('Url of the show',show_info['url'])

# Creating dataframe
df = pd.DataFrame(episodes_info)
# Saving dataframe
df.to_csv('westworld_episodes_info.csv')
```
